```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
import datetime
```

```python
# Load the file
file = ("Resources/taiwan202208.csv")
# Read the file
df = pd.read_csv(file)
df
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'], unit='s')
```

```python
df['Date'] = df['Date'].apply(lambda x: x.date())
```

```python
df = df.set_index('Date')
```

```python
df.index.name = None
```

```python
df
```

```python
df.plot(figsize=(12,10))
plt.xlabel('Date')
plt.ylim(0,800000)
```
